```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
headernames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
figure(figsize=(10, 8), dpi=80)

plt.style.use('ggplot')
plt.xlabel('age and bmi')
plt.ylabel('charges')
plt.scatter(x=df['SepalLengthCm'],y=df['Species'],c='b',s=100,alpha=0.6,marker='o',edgecolors='y')
plt.scatter(x=df['PetalLengthCm'],y=df['Species'],c='r',s=100,alpha=0.5,marker='h',edgecolors='b')
plt.scatter(x=df['PetalWidthCm'],y=df['Species'],s=100,c='k',alpha=0.5,marker='h',edgecolors='b')
plt.scatter(x=df['SepalWidthCm'],y=df['Species'],c='g',s=100,alpha=0.5,marker='h',edgecolors='b')
plt.tight_layout()
plt.show()
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, 5].values
```

```python
print(df.shape)
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
new_data=[[6.1,3.5,5.5,2.2]]
prediction = classifier.predict(new_data)
prediction[0]
```

```python
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```
